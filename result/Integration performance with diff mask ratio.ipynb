```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
```

```python
data = pd.read_csv("figure1_all.csv")
```

```python
data
```

```python
labels = ['10%', '20%', '50%', '100%']

# x = np.arange(len(labels))  # the label locations
x = np.array([0, 0.7, 1.4, 2.1])
width = 0.06  # the width of the bars

fig, axes = plt.subplots(1, 1, figsize=(6, 2))


col = "Overall score"
rects0 = axes.bar(x - width/2, data[(data['Model'] == "[0,0.1]")][col], width, label = '[0,0.1]', color = '#8dd3c7')
rects1 = axes.bar(x + width/2 + 0.02, data[(data['Model'] == "[0,0.3]")][col], width, label = '[0,0.3]', color = '#fc9d62')
rects2 = axes.bar(x + width*3/2 + 0.04, data[(data['Model'] == "[0,0.5]")][col], width, label = '[0,0.5]', color = '#bebada')
rects3 = axes.bar(x + width*5/2 + 0.06, data[(data['Model'] == "[0,0.7]")][col], width, label = '[0,0.7]', color = '#fb8072')
rects4 = axes.bar(x + width*7/2 + 0.08, data[(data['Model'] == "[0,0.9]")][col], width, label = '[0,0.9]', color = '#80b1d3')
rects5 = axes.bar(x + width*9/2 + 0.10, data[(data['Model'] == "Baseline")][col], width, label = 'Baseline', color = '#999999')

axes.bar_label(rects0, padding = 2, fmt='%.3f', fontsize = 8, rotation = 45)
axes.bar_label(rects1, padding = 2, fmt='%.3f', fontsize = 8, rotation = 45)
axes.bar_label(rects2, padding = 2, fmt='%.3f', fontsize = 8, rotation = 45)
axes.bar_label(rects3, padding = 2, fmt='%.3f', fontsize = 8, rotation = 45)
axes.bar_label(rects4, padding = 2, fmt='%.3f', fontsize = 8, rotation = 45)
axes.bar_label(rects5, padding = 2, fmt='%.3f', fontsize = 8, rotation = 45)

axes.tick_params(axis='x', direction = 'in', bottom = False)
axes.tick_params(axis='y', direction = 'out', labelsize = 10, length = 3)
axes.set_xticks(x + 0.1)
axes.set_xticklabels(labels, size = 10)
axes.set_ylim(bottom = 0.8, top = 0.901)
axes.set_yticks(np.arange(0.8, 0.901, step = 0.02))
axes.set_ylabel(col, fontsize = 12)
    

for spine in ['top', 'right']:
    axes.spines[spine].set_color('none')
axes.legend(
    loc='upper center',
    bbox_to_anchor=(0.5, 1.3),  
    ncol=6, 
    fontsize=10,
    frameon=False
)

plt.savefig('figure1_1.svg',dpi = 300, bbox_inches = 'tight')
plt.show()
```
