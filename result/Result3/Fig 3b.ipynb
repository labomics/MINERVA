```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
```

```python
data = pd.read_csv("figure3.csv")
```

```python
data
```

```python
labels = ['Subset 10%', 'Subset 20%', 'Subset 50%', 'Subset 100%']

# x = np.arange(len(labels))  # the label locations
x = np.array([0, 0.6, 1.2, 1.8])
width = 0.08  # the width of the bars

fig, axes = plt.subplots(1, 1, figsize=(7.5, 2.5))

rects0 = axes.bar(x - width/2, data[(data['Model'] == "MINERVA")]['F1 score'], width, label = 'MINERVA', color = '#FA7F6F')
rects1 = axes.bar(x + width/2 + 0.03, data[(data['Model'] == "MINERVA_mask+noise")]['F1 score'], width, label = 'MINERVA_mask+noise', color = '#F5BE81')
rects2 = axes.bar(x + width*3/2 + 0.06, data[(data['Model'] == "MINERVA_mask+subsample")]['F1 score'], width, label = 'MINERVA_mask+subsample', color = '#A0CBEB')
rects3 = axes.bar(x + width*5/2 + 0.09, data[(data['Model'] == "Null model")]['F1 score'], width, label = 'Null model', color = '#999999')

axes.bar_label(rects0, padding = 2, fmt = '%.3f', fontsize = 8, rotation = 45)
axes.bar_label(rects1, padding = 2, fmt = '%.3f', fontsize = 8, rotation = 45)
axes.bar_label(rects2, padding = 2, fmt = '%.3f', fontsize = 8, rotation = 45)
axes.bar_label(rects3, padding = 2, fmt = '%.3f', fontsize = 8, rotation = 45)

axes.tick_params(axis = 'x', direction = 'in', bottom = False)
axes.tick_params(axis = 'y', direction = 'out', labelsize = 10, length = 3)
axes.set_xticks(x + 0.1)
axes.set_xticklabels(labels, size = 10)
axes.set_ylim(bottom = 0.8, top = 0.91)
axes.set_yticks(np.arange(0.8, 0.91, step = 0.05))
axes.set_ylabel("F1_bc score", fontsize = 12)

# axes.set_title("impute 50% of missing gene expression", fontsize = 10, y = 0.9, fontweight = "bold")
axes.legend(loc = 'upper center', ncol = 4, fontsize = 8, frameon = False)

for spine in ['top', 'right']:
    axes.spines[spine].set_color('none')


plt.savefig('figure3.svg',dpi = 300, bbox_inches = 'tight')
plt.show()
```
